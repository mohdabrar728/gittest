# Bar graph

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.bar(x,y)
```

```python
plt.bar(x,y)
plt.title('Bar Chat')
plt.xlabel('Student')
plt.ylabel('Attendance')
```

# Different Data in the Same Bar graph

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
x1 = [1,2,3,4,5]
y1 = [40,50,60,70,80]
plt.bar(x,y,alpha=.4,label='First data',color='r')
plt.bar(x1,y1,alpha=.6,label='Second data')
plt.legend()
```

# Histograms

```python
import numpy as np
import pandas as pd
```

```python
height = np.arange(150,170,2)
height
```

```python
plt.hist(height,rwidth=.4,color='r',bins=4)
```

```python
plt.hist(height,rwidth=.4,color='y',bins=4,label='height')
plt.legend()
```

```python
bin = [150,156,160,168]
plt.hist(height,rwidth=.3,color='g',bins=bin)
```

# Scatter block

```python
x = [1,3,5,7,9]
y = [10,45,65,8,32]
plt.scatter(x,y,marker='*')
```

# Stack/Area Block

```python
month = [1,2,3,4,5]
java = [50,40,25,10,9]
sql = [30,25,20,15,0]
python = [35,18,10,40,15]
plt.stackplot(month,java,sql,python)
```

# PIE chart

```python
activities = ['read','caht','eat','sleep']
slices = [1,5,3,15]
colors= ['g','y','r','lightgreen']
plt.pie(slices,labels=activities,colors= colors)
plt.legend()
```

```python
plt.pie(slices,labels=activities,colors=colors,startangle=90,autopct='%1.1f%%')
plt.legend()
```

```python
temp = pd.read_excel('temp_details.xlsx')
temp
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
x = temp['Day']
y = temp['Temp']

plt.plot(x,y,color='r',linewidth=3)
plt.title('One Week Temp Details')
plt.xlabel('Day')
plt.ylabel('Temp')
```

```python
plt.bar(x,y)
```

```python
plt.stackplot(x,y)
```

```python
plt.scatter(x,y,marker='*',s=100,color='g')
```

```python
plt.pie(y,labels=x,colors=['r','g','y','k','c','m','b'],explode=[0,0,0,0,0,0,0,0,1])
```
